```python
import unirest
import pandas as pd 

# get 20 random recipes
response = unirest.get("https://spoonacular-recipe-food-nutrition-v1.p.mashape.com/recipes/random?limitLicense=false&number=20",
  headers={
    "X-Mashape-Key": "uI9T1GTt8nmshcUUWJOjq8TQNGBgp1P9Zffjsn7dAbkmTSDt1k",
    "Accept": "application/json"
  }
)
```

```python
# get relevant data of the recipes and put it in a dataframe
testData = response.body
testFrame = pd.DataFrame(testData['recipes'], columns=['id', 'title', 'readyInMinutes'])
testFrame
```

```python
recipesNutritions = []
# for each recipe
for recipeId in testFrame['id']:
    url = "https://spoonacular-recipe-food-nutrition-v1.p.mashape.com/recipes/" + str(recipeId) + "/information?includeNutrition=true"
    response = unirest.get(url,
      headers={
        "X-Mashape-Key": "uI9T1GTt8nmshcUUWJOjq8TQNGBgp1P9Zffjsn7dAbkmTSDt1k",
        "Accept": "application/json"
      }
    )
    recipeInfo = response.body
    # get nutrition information
    recipeNutritions = recipeInfo['nutrition']['nutrients']
    # create dict of nutrition and its amount
    nutritionsDict = {nutrition['title']: nutrition['amount'] for nutrition in recipeNutritions}
    # add it to list of recipes nutrients
    recipesNutritions.append(nutritionsDict)
```

```python
# save 
import json
json.dump(recipesNutritions, open("recipesNutrients.txt",'w'))
```
