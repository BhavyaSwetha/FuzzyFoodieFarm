```python
def getCost(stringName): 
    response = unirest.post("https://spoonacular-recipe-food-nutrition-v1.p.mashape.com/recipes/visualizePriceEstimator",
    headers={
    "X-Mashape-Key": "uI9T1GTt8nmshcUUWJOjq8TQNGBgp1P9Zffjsn7dAbkmTSDt1k",
    "Accept": "text/html",
    "Content-Type": "application/x-www-form-urlencoded"
    },
    params={
    "defaultCss": False,
    "ingredientList": stringName,
    "showBacklink": False
    })
    soup = BeautifulSoup(response.body, "lxml") 
    ingredient = soup.find("div", {"class":"spoonacular-quickview"})
    costField = str(ingredient.string)
    index = costField.find('$')
    cost = float(costField[index+1:])
    return cost
```

```python
def getInfo(recipeId): 
    url = "https://spoonacular-recipe-food-nutrition-v1.p.mashape.com/recipes/" + str(recipeId) + "/information?includeNutrition=true"
    info = unirest.get(url,
    headers={
    "X-Mashape-Key": "uI9T1GTt8nmshcUUWJOjq8TQNGBgp1P9Zffjsn7dAbkmTSDt1k",
    "Accept": "application/json"
    })
    time = info.body['readyInMinutes']
    dishtypes = info.body['dishTypes']
    calories = info.body['nutrition']['nutrients'][0]['amount']
    costList = []
    for i in info.body['extendedIngredients']:
        name = i['originalString']
        cost = getCost(name)
        costList.append(cost)
    cost = sum(costList)
    return pd.Series([time, dishtypes, calories, cost], index=["time", "dishtypes", "calories", "cost"])
```

```python
extraData = testFrame['id'].apply(getInfo)
```

```python
recipeData = pd.concat([testFrame, extraData], axis=1)
```
